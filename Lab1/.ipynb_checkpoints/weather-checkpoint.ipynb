```python
#ASHRAF ALI - 19315281 - LAB1 - STATISTICS WITH PYTHON
import pandas as pd
import matplotlib.pyplot as plt
```

```python
weather_data = pd.read_csv('weather2019.csv')
```

```python
#Q1 What was the average wind speed on 19-May-19 (to 1 decimal place)?
print(weather_data.loc[weather_data['date'] == '19-May-19', 'wind'])
```

```python
#Q2 Find the mean hours of sunlight per day (round your answer to 2 decimal places).
print(weather_data['sun'].mean())
```

```python
#Q3 What was the largest amount of rain to fall in one day (round your answer to 1 decimal places)?
print(weather_data['rain'].max())
```

```python
#Q4 Find the median of the daily windspeed (round your answer to 1 decimal places).
print(weather_data['wind'].median())
```

```python
#Q5 Find the standard deviation of the daily rainfall (round your answer to 2 decimal places).
print(weather_data['rain'].std())
```

```python
#Q6 On what date was the largest rainfall recorded (state your answer in day-month-year format, e.g. 21-Apr-97)?
print(weather_data.loc[weather_data['rain'] == weather_data['rain'].max(), 'date'])
```

```python
#Q7 Create a scatter plot of the hours of sunlight per day against the minimum daily temperature. 
#Which of the following best describes the plot?
plt.figure()
plt.scatter(weather_data['sun'],weather_data['mintemp'])
plt.xlabel('Hours of Sunlight per day')
plt.ylabel('Minimum Daily Temperature')
```

```python
#Q8 Create a histogram for the minimum daily temperature. Which of the following best describes the plot?
plt.figure()
plt.hist(weather_data['mintemp'], bins = 20, color='k')
plt.xlabel('Minimum Daily Temperature')
```

```python
#Q9 The two commands below do the same thing.
plt.plot(weather_data.rain,weather_data.mintemp,'o')
```

```python
#Q9 The two commands below do the same thing.
plt.scatter(weather_data.rain,weather_data.mintemp)
```

```python
#Q10 When using the plt.plot() command, which of the following line styles produces alternating dots and dashes?
plt.plot(weather_data.rain,weather_data.mintemp,'-.')
```
